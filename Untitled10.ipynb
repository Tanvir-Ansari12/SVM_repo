```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("data.csv")
df
```

```python
df.drop(["id", "Unnamed: 32"], axis=1, inplace=True, errors='ignore')
```

```python
df.dropna(inplace=True)
```

```python
df["diagnosis"] = LabelEncoder().fit_transform(df["diagnosis"])
```

```python
X = df.drop("diagnosis", axis=1)
y = df["diagnosis"]
```

```python
X = X.dropna(axis=1)
y = y[~X.isnull().any(axis=1)]
X = X[~X.isnull().any(axis=1)]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)
```

```python
svm_linear = SVC(kernel="linear", C=1.0)
svm_linear.fit(X_train, y_train)
y_pred_linear = svm_linear.predict(X_test)
print("Linear SVM:\n", classification_report(y_test, y_pred_linear))
```

```python
svm_rbf = SVC(kernel="rbf", C=1.0, gamma="scale")
svm_rbf.fit(X_train, y_train)
y_pred_rbf = svm_rbf.predict(X_test)
print("RBF SVM:\n", classification_report(y_test, y_pred_rbf))
```

```python
accuracy_linear = accuracy_score(y_test, y_pred_linear)
print("Linear SVM Accuracy:", accuracy_linear)

accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
print("RBF SVM Accuracy:", accuracy_rbf)
```
